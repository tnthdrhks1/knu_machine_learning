# 원주 코로나 예측

```
url = 'http://covid19.wonju.go.kr/intro_new_sub.jsp'
df = pd.read_html(url)
df
```

```
len(df[2]), len(df[3])
```

```
df[2].columns = df[1].columns
df[2]
```

```
df_save = df
df = df[2]
```

```
df.columns
```

```
df['확진자 관리번호'][0] = df['확진자 관리번호'][0].replace('번', '')
```

```
for i in range(len(df)):
  df['확진자 관리번호'][i] = df['확진자 관리번호'][i].replace('번', '').replace('영등포구', '').replace('홍천군', '') \
  .replace('마포구', '').replace('관악구', '').replace('은평구', '').replace('광진구', '').replace('양평군', '').replace('제천시', '')
df
```

```
df['확진자 관리번호'] = df['확진자 관리번호'].astype(np.int)
df = df.rename(columns={'확진자 관리번호':'id'})
df
```

```
df = df.set_index('id')
df
```

```
plt.figure(figsize=(15,5))
df['감염경로(추정)'].hist()
```

```
plt.figure(figsize=(8,5))
df['감염경로(추정)'].value_counts().plot.barh()
```
